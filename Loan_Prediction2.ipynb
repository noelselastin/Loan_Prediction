```python
#Data Wrangling
```

```python
#Removing missing values
```

```python
from sklearn.impute import SimpleImputer
 
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
cols=['Gender','Married','Self_Employed','Dependents','Loan_Amount_Term','Credit_History']
train[cols] = imputer.fit_transform(train[cols])
```

```python
train[['LoanAmount','Loan_Amount_Term','Credit_History']].nunique()
```

```python
train.nunique()
```

```python
train.dtypes
```

```python
for i in ['LoanAmount']:
    train[i]=train[i].fillna(train[i].median())
```

```python
train.isnull().sum()
```

```python
train=train.drop('Loan_ID',axis=1)
```

```python
#One hot encoding
```

```python
h_columns=list(train) 
h_columns.remove('ApplicantIncome') 
h_columns.remove('CoapplicantIncome') 
h_columns.remove('LoanAmount') 
h_columns.remove('Loan_Status') 
h_columns.remove('Loan_Amount_Term')
```

```python
train2=pd.get_dummies(train,columns=h_columns)
```

```python
train2.shape
```

```python
train2.columns
```

```python
pd.set_option("display.max.rows",None)
pd.set_option("display.max.columns",None)
train2.dtypes
```

```python
train2['Loan_Amount_Term'] = train2['Loan_Amount_Term'].astype(int)
```

```python
train2.dtypes
```

```python
#Label Encoding
```

```python
train2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
train2['Loan_Status']=label_encoder.fit_transform(train2['Loan_Status'])
```

```python
train2.head()
```

```python
#Data Analysis
#Cross Validation
```

```python
X=train2.drop('Loan_Status',axis=1)
y=train2['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lm=LogisticRegression()
lm.fit(X_train,y_train)
y_pred=lm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print("Accuracy=",accuracy_score(y_test,y_pred))
print("Precision=",precision_score(y_test,y_pred))
print("Recall=",recall_score(y_test,y_pred))
print("f1 score=",f1_score(y_test,y_pred))
```
